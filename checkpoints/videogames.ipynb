```python
# import libraries for data preprocessing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# read the csv file
df = pd.read_csv('Video-games.csv')
```

```python
# info about the dataset
df.info()
```

```python
# sum the null values in each column
df.isnull().sum()
```

```python
# replace the missing values with the mean of the year column
df['Year'].fillna(df['Year'].mean(), inplace=True)
```

```python
# print the rows where the publisher column is null
df[df['Publisher'].isnull()]
```

```python
# drop rows that has null values in the publisher column
df.dropna(subset=['Publisher'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
# an area chart showing the total sales of each year
```

```python
# Create an area chart to show the trend of global sales over the years
df.groupby('Year')['Global_Sales'].sum().plot(kind='area', figsize=(15, 5))
plt.title('Global Sales over the years')
plt.ylabel('Global Sales')
plt.xlabel('Year')
plt.show()
```

```python
ax2 = sns.lineplot(data=df, x='Year', y='Global_Sales', estimator=None, lw=2)

# Set title and labels
ax2.set_title('Area Chart using Seaborn')
ax2.set_xlabel('Year')
ax2.set_ylabel('Global Sales')

# Show the chart
plt.show()
```

```python
fig, ax3 = plt.subplots()

# Fill the area under the curve
ax3.fill_between(df['Year'], df['Global_Sales'], color='blue', alpha=0.2)

# Draw the line
ax3.plot(df['Year'], df['Global_Sales'], color='blue', lw=2)

# Set title and labels
ax3.set_title('Area Chart using Matplotlib')
ax3.set_xlabel('Year')
ax3.set_ylabel('Global Sales')

# Show the chart
plt.show()
```
